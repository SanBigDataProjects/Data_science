Potential Limitations and Assumptions of ANOVA and Linear Models

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor
import scipy as sp

# Load the data
data = pd.read_csv("data.csv")
print(data.head())

# Assuming the last column is the target
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

```

```python
# Check for missing values
print("Missing values in X:\n", X.isnull().sum())
print("Missing values in y:\n", y.isnull().sum())

# Drop rows with missing values in X or y
df = X.copy()
df['target'] = y
df = df.dropna()

# Extract aligned X and y
X_aligned = df.drop(columns=['target']).values
y_aligned = df['target'].values
print(f"New Length of X: {len(X_aligned)}")
print(f"New Length of y: {len(y_aligned)}")
```

```python
# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X_aligned, y_aligned, test_size=0.3, random_state=1)

# Checking the shapes of the splits
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")

# Training the model
model = LinearRegression()
model.fit(X_train, y_train)
```

1. Linear Relationship

```python
# Ensure the column names match those in your data
column_names = data.columns[:-1]

# Plotting features vs target
fig, axes = plt.subplots(ncols=3, figsize=(12, 2.5))
for i, ax in enumerate(axes):
    ax.scatter(df[column_names[i]], df['target'])
    ax.set_title(f"Feature {i+1}")

plt.show()
```

2. Multicollinearity

```python
# Calculate Variance Inflation Factor (VIF) for multicollinearity check
vif = [variance_inflation_factor(X_train, i) for i in range(X_train.shape[1])]
print(pd.DataFrame({'VIF': vif}, index=column_names))
```

```python
# Correlation heatmap
sns.heatmap(df.iloc[:,0:3].corr(), annot=True)
```

3. Normality of Residual

```python
# Residuals analysis
residuals = y_train - model.predict(X_train)

# Distribution plot of residuals
sns.displot(residuals, kind='kde')
plt.title('Residuals Distribution')
plt.show()
```

```python
# QQ Plot for normality of residuals
fig, ax = plt.subplots(figsize=(6,4))
sp.stats.probplot(residuals, plot=ax, fit=True)
plt.title('QQ Plot')
plt.show()
```

4. Homoscedasticity

```python
# Plot residuals vs predicted values
y_pred = model.predict(X_train)
plt.scatter(y_pred, residuals)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted Values')
plt.show()
```

5. Autocorrelation of Residuals

```python
# Plotting residuals over the order of observations
plt.plot(residuals)
plt.title('Residuals Over Time')
plt.xlabel('Observation Order')
plt.ylabel('Residual')
plt.show()
```

Assumption of Anova 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Generate sample data
np.random.seed(0)
data = pd.DataFrame({
    'feature1': np.random.normal(0, 1, 100),
    'feature2': np.random.normal(5, 2, 100),
    'feature3': np.random.normal(10, 3, 100),
    'target': np.random.normal(50, 10, 100)
})
```

```python
# Create categorical groups from one of the features (e.g., feature1)
data['feature1_cat'] = pd.cut(data['feature1'], bins=3, labels=['low', 'medium', 'high'])

# Check for normality using QQ plots
for group in data['feature1_cat'].unique():
    stats.probplot(data[data['feature1_cat'] == group]['target'], dist="norm", plot=plt)
    plt.title(f"QQ plot for feature1 category {group}")
    plt.show()
```

```python
# Check for homogeneity of variances using Levene's test
levene_test = stats.levene(
    *[data[data['feature1_cat'] == group]['target'] for group in data['feature1_cat'].unique()]
)
print(f"Levene's test for homogeneity of variances: p-value = {levene_test.pvalue}")

# Perform ANOVA
model = ols('target ~ C(feature1_cat)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
# If ANOVA is significant, perform post-hoc tests
if anova_table['PR(>F)'].iloc[0] < 0.05:
    tukey = pairwise_tukeyhsd(endog=data['target'], groups=data['feature1_cat'], alpha=0.05)
    print(tukey.summary())

    # Plotting the result
    tukey.plot_simultaneous(comparison_name='low')
    plt.show()

# Visualization of the data
sns.boxplot(x='feature1_cat', y='target', data=data)
plt.title("Boxplot of target by feature1 categories")
plt.show()
```
