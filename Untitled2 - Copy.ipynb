```python
import numpy as np
import matplotlib.pyplot as plt


# Dataset
dataset = [11, 10, 12, 14, 12, 15, 14, 13, 15, 102, 12, 14, 17, 19, 107, 10, 13, 12, 14, 12, 108, 12, 11, 14, 13, 15, 12, 10, 14, 13, 15, 10]



#using z-score method
def detect_outlier(data):
    outliers = []  # Define inside the function to avoid global side effects
    threshold = 2.5  # 3rd standard deviation
    mean = np.mean(data)
    std = np.std(data)
    print("Mean:", mean)
    print("Standard Deviation:", std)

    for i in data:
        z_score = (i - mean) / std
        print("Data point:", i, "Z-score:", z_score)
        if np.abs(z_score) > threshold:
            outliers.append(i)
    
    return outliers


# Detect outliers using Z-score method
outliers_zscore = detect_outlier(dataset)
print("Outliers (Z-score method):", outliers_zscore)

# Remove outliers using Z-score method
dataset_no_outliers_zscore = [x for x in dataset if x not in outliers_zscore]

# Plot the data before and after removing outliers using Z-score method
plt.figure(figsize=(10, 6))
plt.hist(dataset, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Data with Outliers (Z-score method)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(dataset_no_outliers_zscore, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Data without Outliers (Z-score method)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# IQR method to detect outliers
def detect_outlier_iqr(data):
    outliers = []
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    for i in data:
        if i < lower_bound or i > upper_bound:
            outliers.append(i)
    return outliers
# Detect outliers using IQR method
outliers_iqr = detect_outlier_iqr(dataset)
print("Outliers (IQR method):", outliers_iqr)
```

```python
# Remove outliers using IQR method
dataset_no_outliers_iqr = [x for x in dataset if x not in outliers_iqr]

# Plot the data before and after removing outliers using IQR method
plt.figure(figsize=(10, 6))
plt.hist(dataset, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Data with Outliers (IQR method)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(dataset_no_outliers_iqr, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Data without Outliers (IQR method)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
