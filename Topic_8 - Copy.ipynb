```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("Titanic.csv")
print(data.head())


# Display the number of missing values for each column
print(data.isnull().sum())

# Fill missing values in 'Age' with the median age
data['Age'] = data['Age'].fillna(data['Age'].median())

# Fill missing values in 'Embarked' with the mode
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])

# Fill missing values in 'Cabin' with 'M' and create 'Deck' feature
data['Cabin'] = data['Cabin'].fillna('M')
data['Deck'] = data['Cabin'].apply(lambda x: x[0])

# Drop the original 'Cabin' column
data.drop(columns=['Cabin'], inplace=True)

# Display summary statistics
summary = data.describe()
print(summary)
```

```python
# Univariate Analysis: Count plot for survival count
sns.countplot(x='Survived', data=data)
plt.title("Survival Count")
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Univariate Analysis: Histogram for Age distribution
sns.histplot(data['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bivariate Analysis: Bar plot for survival rate by gender
sns.barplot(x='Sex', y='Survived', data=data)
plt.title('Survival Rate by Gender')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Bivariate Analysis: Bar plot for survival rate by passenger class
sns.barplot(x='Pclass', y='Survived', data=data)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Bivariate Analysis: Scatter plot for age vs. fare, colored by survival
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=data)
plt.title('Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived')
plt.show()
```

```python
# Multivariate Analysis: Pairplot to visualize relationships between multiple variables
sns.pairplot(data, hue='Survived', vars=['Age', 'Fare', 'Pclass', 'SibSp', 'Parch'])
plt.show()
```
