```python
x=np.arange(1,11)
```

```python
y1=x*2
y2=x**2
x
y1
y2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='r', linestyle=':',linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y2,color='g',linewidth=4)
plt.grid(True)
plt.show()
```

```python
plt.plot(x,y1,color='y', linewidth=4)
plt.plot(x,y2,color='r',linewidth=3,linestyle=":")
plt.grid(True)
plt.show()

plt.subplot(2,1,1)
plt.plot(x,y1, linestyle=":", color="g")
plt.grid(True)
plt.show()

plt.subplot(2,1,2)
plt.plot(x,y2,color='y',linewidth=4)
plt.grid(True)
plt.show()
```

```python
#Bar Chart

Student={"Sam":67,"San":78,"Emma":67}
names=list(Student.keys())
marks=list(Student.values())

plt.bar(names,marks, linestyle=":",color="r")
plt.title("Student _bar plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.show()
```

```python
#Bar Chart

Student={"Sam":67,"San":78,"Emma":67}
names=list(Student.keys())
marks=list(Student.values())

plt.barh(names,marks, linestyle=":",color="y")
plt.title("Student _bar plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.show()
```
