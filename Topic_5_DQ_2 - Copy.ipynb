```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

# Example data
np.random.seed(0)
X = 2 - 3 * np.random.normal(0, 1, 100)
y = X - 2 * (X ** 2) + np.random.normal(-3, 3, 100)

# Reshape X for sklearn
X = X[:, np.newaxis]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Fit a linear regression model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
y_pred_linear = linear_model.predict(X_test)

# Check residuals
plt.scatter(X_test, y_test - y_pred_linear)
plt.title('Residual plot for Linear Regression')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.show()

# Polynomial regression (to address non-linearity)
polynomial_features = PolynomialFeatures(degree=2)
X_poly = polynomial_features.fit_transform(X_train)

# Fit polynomial regression model
poly_model = LinearRegression()
poly_model.fit(X_poly, y_train)
y_pred_poly = poly_model.predict(polynomial_features.transform(X_test))

# Check residuals again
plt.scatter(X_test, y_test - y_pred_poly)
plt.title('Residual plot for Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.show()
```
