```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load the data
data = pd.read_csv('waterquality.csv')

# Selecting numeric columns for simplicity
numeric_data = data[['Salinity (ppt)', 'DissolvedOxygen (mg/L)', 'pH', 'SecchiDepth (m)', 
                     'WaterDepth (m)', 'WaterTemp (C)', 'AirTemp (C)']]

# Handling missing values (dropna for simplicity)
numeric_data = numeric_data.dropna()

# Assuming 'Salinity (ppt)' is your dependent variable
y = numeric_data['Salinity (ppt)']
X = numeric_data.drop(columns=['Salinity (ppt)'])

# Add constant to X for intercept in the model
X = sm.add_constant(X)

# Fit a linear regression model
model = sm.OLS(y, X).fit()

# Get fitted values and residuals
fitted = model.fittedvalues
residuals = model.resid

# Plot residuals against fitted values
plt.figure(figsize=(8, 6))
sns.residplot(x=fitted, y=residuals, lowess=True)
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()

# Summary of the linear model
print(model.summary())

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Step 2: Generate synthetic data (nonlinear relationship)
np.random.seed(0)
X = np.random.uniform(low=-3, high=3, size=100)
y = X**2 + np.random.normal(loc=0, scale=1, size=100)

X = X[:, np.newaxis]  # Reshape X to fit sklearn API

# Step 3: Fit polynomial regression model
poly_features = PolynomialFeatures(degree=2)  # Degree 2 for quadratic relationship
X_poly = poly_features.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

# Step 4: Visualize the results
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression Example')
plt.grid(True)

# Sort X for better visualization of the polynomial curve
X_sorted = np.linspace(X.min(), X.max(), 100)
X_sorted = X_sorted[:, np.newaxis]
y_pred = model.predict(poly_features.transform(X_sorted))

plt.plot(X_sorted, y_pred, color='red', linewidth=2, label='Polynomial Fit')
plt.legend()
plt.show()

# Print coefficients of the polynomial model
print(f'Polynomial coefficients: {model.coef_}')
print(f'Intercept: {model.intercept_}')
```
