```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url="https://raw.githubusercontent.com/SanBigDataProjects/Data_science/main/placement.csv"
data=pd.read_csv(url)
data.head()
```

```python
plt.scatter(data["cgpa"],data["package"])
plt.xlabel("cgpa")
plt.ylabel("package")
plt.show()
```

```python
data_input=data.iloc[:,0:1]
data_output=data.iloc[:,1:]
data_output
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(data_input,data_output,test_size=0.2,random_state=2)
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
x_test.iloc[0].values
```

```python
lr.predict(x_test.iloc[2].values.reshape(1,1))
```

```python
plt.scatter(data["cgpa"],data["package"])
plt.plot(x_test,lr.predict(x_test),color="red")
plt.xlabel("cgpa")
plt.ylabel("package")
plt.show()
```
