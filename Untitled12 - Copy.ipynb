```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy import stats
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Load the dataset
data_path = 'waterquality.csv'  # Update this with the correct path
environmental_data = pd.read_csv(data_path)

# Clean column names to remove spaces and special characters
environmental_data.columns = environmental_data.columns.str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

# Display the first few rows of the dataset
print(environmental_data.head())

# Visualizing distributions and checking normality
sns.histplot(environmental_data['WaterTemp_C'], kde=True)
plt.title('Distribution of Water Temperature')
plt.show()

# Q-Q plot for normality check
sm.qqplot(environmental_data['WaterTemp_C'], line ='45')
plt.title('Q-Q Plot of Water Temperature')
plt.show()

# Levene's test for homogeneity of variances
stat, p = stats.levene(environmental_data['Salinity_ppt'], environmental_data['DissolvedOxygen_mgL'])
print('Levenes test statistic:', stat)
print('Levenes test p-value:', p)

# Linear regression model
model = ols('WaterTemp_C ~ Salinity_ppt + DissolvedOxygen_mgL + pH', data=environmental_data).fit()
print(model.summary())

# Checking multicollinearity
X = environmental_data[['Salinity_ppt', 'DissolvedOxygen_mgL', 'pH']]
X = sm.add_constant(X)
vif = pd.DataFrame()
vif["Variables"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print(vif)


```
