```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

# Example dataset
data = {'Experience': [5, 7, 3, 10, 8, 6, 2, 12, 9, 4],
        'Department': ['Sales', 'HR', 'IT', 'Sales', 'IT', 'HR', 'IT', 'Sales', 'HR', 'IT'],
        'Salary': [60000, 70000, 50000, 90000, 80000, 65000, 45000, 95000, 75000, 52000]}
df = pd.DataFrame(data)

# Check data types
print(df.dtypes)

# Convert categorical variable 'Department' using one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Department'], drop_first=True)

# Check for missing values
print(df_encoded.isnull().sum())

# Ensure all data is numeric and no missing values
assert df_encoded.apply(lambda col: col.map(lambda x: isinstance(x, (int, float))).all()).all(), "Non-numeric data found in DataFrame"

```

```python
# Exploratory Data Analysis (EDA)
plt.figure(figsize=(4, 4))
plt.scatter(df["Experience"], df["Salary"], color="red")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.boxplot(x='Department', y='Salary', data=df, hue='Department', palette="coolwarm", legend=False)
plt.xlabel("Department")
plt.ylabel("Salary")
plt.show()

# Prepare data for regression
input = df_encoded[['Experience', 'Department_IT', 'Department_Sales']]
output = df_encoded['Salary']

# Convert data to numpy array to ensure correct format
input = np.asarray(input, dtype=float)
output = np.asarray(output, dtype=float)

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(input, output, test_size=0.2, random_state=42)

# Linear Regression using sklearn
lr = LinearRegression()
lr.fit(x_train, y_train)

print("Coefficients:", lr.coef_)
print("Intercept:", lr.intercept_)

# Linear Regression using statsmodels for detailed summary
X = sm.add_constant(input)  # Adds a constant term to the predictor
model = sm.OLS(output, X).fit()
print(model.summary())
```
