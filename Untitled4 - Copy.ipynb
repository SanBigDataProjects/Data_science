```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p = 0.7  # True probability of heads
n = 1000  # Number of coin flips
alpha = 0.05  # Significance level
z = 1.96  # z-score for 95% confidence

# Simulate coin flips
flips = np.random.binomial(1, p, n)
estimated_p = np.mean(flips)

# Confidence interval
margin_of_error = z * np.sqrt((estimated_p * (1 - estimated_p)) / n)
confidence_interval = (estimated_p - margin_of_error, estimated_p + margin_of_error)

# Visualization
plt.hist(flips, bins=2, edgecolor='black')
plt.xticks([0, 1], ['Tails', 'Heads'])
plt.title('Histogram of Coin Flips')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.show()

# Print results
print(f"Estimated probability of heads: {estimated_p}")
print(f"95% Confidence interval: {confidence_interval}")

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p_true = 0.6  # True probability of heads
n_flips = 1000  # Number of coin flips

# Simulate coin flips
flips = np.random.binomial(1, p_true, n_flips)

# Calculate proportion of heads
p_estimate = np.mean(flips)

# Visualize results
plt.hist(flips, bins=2, edgecolor='black')
plt.xticks([0, 1], ['Tails', 'Heads'])
plt.title('Histogram of Coin Flips')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.show()

# Print results
print(f"Estimated probability of heads: {p_estimate}")
```
