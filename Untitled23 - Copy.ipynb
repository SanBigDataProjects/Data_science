```python
# 3. Consistency Check for dates
data['date'] = pd.to_datetime(data['date'], errors='coerce')
print("Inconsistent dates:\n", data[data['date'].isnull()])

# Drop rows with inconsistent dates
data = data.dropna(subset=['date'])

# Consistency Check for age (remove rows where age is not a number)
data = data[pd.to_numeric(data['age'], errors='coerce').notnull()]
```

```python
# 4. Descriptive Statistics
print("Descriptive statistics:\n", data.describe())

# Visualize distributions of numeric columns
numeric_columns = ['salary', 'years_experience', 'age']
data[numeric_columns].hist(bins=20, figsize=(12, 8))
plt.suptitle('Histograms of Numeric Features')
plt.show()
```

```python

# 5. Correlation Analysis
numeric_data = data[numeric_columns]  # Select only numeric columns for correlation
correlation_matrix = numeric_data.corr()
print("Correlation matrix:\n", correlation_matrix)

# Visualize Correlations
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Save the cleaned data
data.to_csv('cleaned_employee_salaries.csv', index=False)
```
