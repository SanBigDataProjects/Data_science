```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample COVID-19 data
data = pd.DataFrame({
    'Age_Group': ['0-17', '18-49', '50-64', '65+'],
    'Cases': [50000, 200000, 150000, 100000],
    'Deaths': [50, 2000, 5000, 20000]
})

# Calculate mortality rate
data['Mortality_Rate'] = data['Deaths'] / data['Cases'] * 100
print(data)

# Misleading Visualization (Selective Reporting)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.barplot(x='Age_Group', y='Mortality_Rate', data=data)
plt.title("Misleading Mortality Rate")
plt.ylim(0, 1)  # Only showing low mortality rates

# Correct Visualization
plt.subplot(1, 2, 2)
sns.barplot(x='Age_Group', y='Mortality_Rate', data=data)
plt.title("Correct Reporting of Mortality Rate")
plt.ylim(0, 20)  # Showing the actual range

plt.tight_layout()
plt.show()

# Display detailed statistics
print(data)
```
