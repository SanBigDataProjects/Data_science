```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a hypothetical dataset
np.random.seed(0)
data = pd.DataFrame({
    'Teaching_Method': np.repeat(['A', 'B', 'C'], 30),
    'Test_Scores': np.concatenate([
        np.random.normal(75, 10, 30),
        np.random.normal(80, 10, 30),
        np.random.normal(85, 10, 30)
    ])
})

# Perform ANOVA
anova_results = stats.f_oneway(
    data[data['Teaching_Method'] == 'A']['Test_Scores'],
    data[data['Teaching_Method'] == 'B']['Test_Scores'],
    data[data['Teaching_Method'] == 'C']['Test_Scores']
)

# Results
print("ANOVA F-value:", anova_results.statistic)
print("ANOVA p-value:", anova_results.pvalue)

# Visualization
plt.figure(figsize=(10, 6))
sns.boxplot(x='Teaching_Method', y='Test_Scores', data=data)
plt.title('Test Scores by Teaching Method')
plt.xlabel('Teaching Method')
plt.ylabel('Test Scores')
plt.show()
```
