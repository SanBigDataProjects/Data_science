```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
gtd_df = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='latin1')

# Initial Exploration
print(gtd_df.head())
print(gtd_df.info())
print(gtd_df.describe())

```

```python
# Univariate Analysis
# Number of attacks by year
sns.histplot(gtd_df['iyear'], bins=30, kde=False)
plt.title('Number of Terrorist Attacks by Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()

```

```python
# Number of fatalities
sns.histplot(gtd_df['nkill'].dropna(), bins=30, kde=False)
plt.title('Distribution of Number of Fatalities')
plt.xlabel('Number of Fatalities')
plt.ylabel('Frequency')
plt.show()

```

```python
# Bivariate Analysis
# Attack type by region
sns.countplot(y='attacktype1_txt', hue='region_txt', data=gtd_df)
plt.title('Attack Type by Region')
plt.xlabel('Count')
plt.ylabel('Attack Type')
plt.show()

```

```python
# Number of fatalities by attack type
sns.boxplot(x='attacktype1_txt', y='nkill', data=gtd_df)
plt.title('Number of Fatalities by Attack Type')
plt.xlabel('Attack Type')
plt.ylabel('Number of Fatalities')
plt.xticks(rotation=90)
plt.show()

```

```python
# Multivariate Analysis
# Attack type, region, and number of fatalities
sns.catplot(x='region_txt', y='nkill', hue='attacktype1_txt', kind='bar', data=gtd_df)
plt.title('Number of Fatalities by Region and Attack Type')
plt.xlabel('Region')
plt.ylabel('Number of Fatalities')
plt.xticks(rotation=90)
plt.show()

```

```python
# Geospatial Analysis
# Using Folium or other libraries to visualize geographic data
import folium

# Create a map centered around the median coordinates
map_center = [gtd_df['latitude'].median(), gtd_df['longitude'].median()]
mymap = folium.Map(location=map_center, zoom_start=2)

# Add points to the map
for idx, row in gtd_df.iterrows():
    if pd.notnull(row['latitude']) and pd.notnull(row['longitude']):
        folium.CircleMarker(location=[row['latitude'], row['longitude']],
                            radius=2,
                            color='red',
                            fill=True).add_to(mymap)

# Save map to an HTML file
mymap.save('gtd_map.html')

# Display the map
mymap

# Time Series Analysis
# Number of attacks over time
gtd_df['date'] = pd.to_datetime(gtd_df[['iyear', 'imonth', 'iday']].astype(str).agg('-'.join, axis=1), errors='coerce')
gtd_df.set_index('date', inplace=True)
gtd_df['iyear'].resample('Y').size().plot()
plt.title('Number of Terrorist Attacks Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()

# Correlation Analysis
# Correlation matrix of numerical variables
correlation_matrix = gtd_df[['nkill', 'nwound', 'property', 'propvalue']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Handling Missing Data
# Check for missing values
missing_data = gtd_df.isnull().sum()
print(missing_data[missing_data > 0])

# Example of filling missing values
gtd_df['nkill'].fillna(0, inplace=True)
gtd_df['nwound'].fillna(0, inplace=True)
```
