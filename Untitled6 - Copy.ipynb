```python
import matplotlib.pyplot as plt

# Example data
years = [2018, 2019, 2020, 2021, 2022]
sales = [50, 55, 53, 60, 70]

plt.figure(figsize=(10, 6))
plt.bar(years, sales)
plt.ylim(0, max(sales) + 10)  # Start y-axis at 0
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Over Years')
plt.show()
```
