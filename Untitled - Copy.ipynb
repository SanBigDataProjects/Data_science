```python
df_country=pd.read_excel('Country-Code.xlsx',engine='openpyxl')
df_country.head()
```

```python
final_df=pd.merge(data,df_country, on='Country Code', how='left')
final_df.head(2)
```

```python
country_names=final_df.Country.value_counts().index
country_value=final_df.Country.value_counts().values
```

```python
plt.pie(country_value[:3], labels=country_names[:3])
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Ratings_counts'})
ratings
```

```python
#import matplotlib
plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Ratings_counts",data=ratings,hue='Rating color')
```

```python
#Find the country names that has given 0 ratings
filtered_df = final_df.groupby('Country').filter(lambda x: (x['Aggregate rating'] == 0).all())
filtered_df
```
