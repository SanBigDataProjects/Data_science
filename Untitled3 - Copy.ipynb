```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
data=pd.read_csv('waterquality.csv')
data.describe()
```

```python
print(data.dtypes)
```

```python
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
numeric_columns
```

```python
for col in numeric_columns:
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    sns.histplot(data[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    

    #plt.show()
```

```python
# Identify outliers using Z-score
z_scores = stats.zscore(data[numeric_columns])
outliers_z = (z_scores > 3) | (z_scores < -3)
outliers_z = outliers_z.any(axis=1)
print("Rows with outliers detected by Z-score:")
#print(data[outliers_z])
```

```python
# Identify outliers using IQR
Q1 = data[numeric_columns].quantile(0.25)
Q3 = data[numeric_columns].quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = ((data[numeric_columns] < (Q1 - 1.5 * IQR)) | (data[numeric_columns] > (Q3 + 1.5 * IQR)))
outliers_iqr = outliers_iqr.any(axis=1)
print("Rows with outliers detected by IQR:")
#print(data[outliers_iqr])
```

```python
# Handling outliers by removing them
df_no_outliers_z = data[~outliers_z]
df_no_outliers_iqr = data[~outliers_iqr]

# Visualizing the impact of handling outliers
for col in numeric_columns:
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 3, 1)
    sns.boxplot(x=data[col])
    plt.title(f'Original {col}')
    
    plt.subplot(1, 3, 2)
    sns.boxplot(x=df_no_outliers_z[col])
    plt.title(f'No Outliers (Z-score) {col}')
    
    plt.subplot(1, 3, 3)
    sns.boxplot(x=df_no_outliers_iqr[col])
    plt.title(f'No Outliers (IQR) {col}')
    
    #plt.show()
```
