```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns

```

```python
data= pd.read_csv("Social_Network_Ads.csv")
data.drop(["User ID","Gender"],axis=1, inplace=True)
```

```python
data.sample(5)
```

```python
data.head()
```

```python
x_train, x_test, y_train,y_test=train_test_split(data.drop("Purchased",axis=1),data["Purchased"],test_size=0.3,random_state=0)
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)
```

```python
scaler.mean_
```

```python
x_traintransfor=scaler.transform(x_train)
x_testtransform=scaler.transform(x_test)
```

```python
x_traintransfor.shape[0]
x_testtransform.shape[0]
```

```python
x_train
```

```python
x_train_scaled=pd.DataFrame(x_traintransfor,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_testtransform,columns=x_test.columns)
x_train_scaled
x_test_scaled
```

```python
np.round(x_train.describe(),1)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(x_train['Age'], x_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(x_train_scaled['Age'], x_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Age'], ax=ax1)
sns.kdeplot(x_train['EstimatedSalary'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(x_train_scaled['Age'], ax=ax2)
sns.kdeplot(x_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```
